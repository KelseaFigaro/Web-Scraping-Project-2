### import needed packages

```python
import pandas as pd

from matplotlib import pyplot as plt

import seaborn as sns
```

### read in the csv files

```python
Gotham= pd.read_csv('GothamPlayers.csv')
```

```python
Spirit= pd.read_csv('SpiritPlayers.csv')
```

```python
Racing= pd.read_csv('RacingPlayers.csv')
```

```python
Thorns= pd.read_csv('ThornsPlayers.csv')
```

## Questions to be answered

### 1) Is there correlation between player time on the field and goals for Gotham FC
### 2) Is there a correlation between passing distance and passing completed for the Washington Spirit?
### 3) What is the most common Age for the Racing Louisville team?
### 4) Is there a correlation between players shot totals and goals for the Thorns?


## Question 1 
### Is there correlation between player time (minutes played) and goals for Gotham FC

```python
sns.scatterplot(x='goals',
                y='minutes',
                data=Gotham,
               color='skyblue')
plt.title("Coorelation Between Minutes and Goals for Gotham FC")
```

## Analysis for Question 1
### Looking at this plot, it looks like there is no coorlation between how many minutes players play versus how many goals they score. This might have to do with the players positions. Goalkeepers are extremely rare occurances that they score goals. Defenders also tend to score less than midfielders and forwards. 

## Question 2
### Is there a correlation between passing distance and passing completed for the Washington Spirit?

```python
sns.regplot(data=Spirit, x='passing distance', y='passing completed', color='black')
plt.title('Coorelation Between Passes Completed And Passing Distance for the Washington Spirit')
```

## Analysis for Question 2
### Looking at this plot, there is postive coorelation betweenn passing completed and passing distance. The total more passing distance players have, the more passes the players have done, meaning that they have completed more passes. Based on the shaded regions of the plot, data has a high confidence interval.

## Question 3
### What is the most common Age for the Racing Louisville team?

#### I used this link to help with this histogram https://www.youtube.com/watch?v=Bjz00ygERxY&t=116s

```python
sns.histplot(x='Player Age', data=Racing, bins=10, stat='count',color='mediumpurple')
plt.title('Most Common Player Age for Racing Louisville FC')
```

## Question 3 Analysis
### Looking at this histogram, it looks like the most common age is around 28. Least common are players who are 18-19 and 26-27. There are currently no players that are 20-21 and 30. 

## Question 4 
### Is there a correlation between players shot totals and goals for the Thorns?

```python
sns.regplot(data=Thorns, x='goals', y='shots',color='indianred')
plt.title('Coorelation Between Player Shot Totals And Goals for the Portland Thorns')
```

## Question 4 Analysis
### There appears to be a postive correlation between shots and goals for the Portland Thorns. As players take more shots, it tends to lead towards more goals. However, as more shots are taken, the confidence inverval becomes lower. This suggestions that there is not enough players getting that many goals and that many shots for there to be higher confidence in the data
